```python
# 3. Scatter Plot: Relationship between Kilometers Driven and Price
plt.figure()
plt.scatter(df['kms_driven'], df['Price'], color='green', alpha=0.7)
plt.title('Kilometers Driven vs Price', fontsize=16)
plt.xlabel('Kilometers Driven', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```

```python
# 4. Box Plot: Price distribution by Fuel Type
plt.figure()
sns.boxplot(x='fuel_type', y='Price', data=df, palette='pastel')
plt.title('Price Distribution by Fuel Type', fontsize=16)
plt.xlabel('Fuel Type', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```

```python
# 5. Line Plot: Price trend over Years
plt.figure()
sns.lineplot(x='year', y='Price', data=df, marker='o', color='blue')
plt.title('Price Trend Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```

```python
# 6. Swarm Plot: Price distribution by Company
plt.figure(figsize=(15,5))
sns.swarmplot(x='company', y='Price', data=df, palette='muted')
plt.title('Price Distribution by Company', fontsize=16)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
